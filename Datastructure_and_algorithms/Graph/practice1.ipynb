# Graph 
- Breadth first search 
- Depth first search 

```python
from collections import defaultdict,deque
class Graph:
    def __init__(self):
        self.Graph = defaultdict(list) 
    
    def add_edge(self,node1,node2):
        self.Graph[node1].append((node2)) 
        return self.Graph

def display_graph(Graph):
    for node in Graph:
        print(f"{node} -> {Graph[node]}")
```

```python
def Breadth_first_search(graph,source):
    seen = set()
    queue = deque([source]) 

    while queue:
        curr_node = queue.popleft() 
        if curr_node in seen:
            continue 
        seen.add(curr_node)
        print(curr_node)
        for edges in graph[curr_node]:
            queue.append(edges) 
```

```python
def Depth_first_search(graph,source):
    seen = set()
    stack = [source] 

    while stack:
        curr_node = stack.pop() 
        if curr_node in seen:
            continue
        seen.add(curr_node)
        print(curr_node) 

        for edges in graph[curr_node]:
            stack.append(edges)
```

```python
graph = Graph()
graph.add_edge(0,1) 
graph.add_edge(0,2) 
graph.add_edge(1,2)
graph.add_edge(2,3) 
graph.add_edge(3,4) 
graph = graph.add_edge(4,5)
print("Original Graph")
display_graph(graph)
print("Breadth first search : ")
Breadth_first_search(graph,0)
print("Depth first search : ")
Depth_first_search(graph,0) 
```

**Objective**: Determine if two nodes in a network are connected.

**Scenario**: Social Network Connectivity
You are given a list of friendships in a small social network. Each friendship is an undirected edge between two people.

**Your Task**: Given a source person and a destination person, write a function to determine if they belong to the same social circle (i.e., if there is a path of mutual friends connecting them). Return True if a path exists, and False otherwise.

```python
class Graph:
    def __init__(self):
        self.Graph = defaultdict(list) 
    
    def add_edge(self,node1,node2):
        self.Graph[node1].append(node2) 
        self.Graph[node2].append(node1)
    
    def save_graph(self):
        return self.Graph

def display_graph(Graph):
    for node in Graph:
        print(f"{node} -> {Graph[node]}")
```

```python
def Dfs_friends(graph,src,des):
    seen = set() 
    stack = [source] 

    while stack:
        curr_node = stack.pop() 
        
        if curr_node == des:
            print("they belong to same social circle")
            return

        if curr_node in seen:
            continue

        seen.add(curr_node) 


        for edge in graph[curr_node]:
            if edge not in seen:
                stack.append(edge)

    print("They do not belong in same social circle")
```

```python
friendship_circle = [[0, 1], [0, 2],[3,0], [3, 5], [5, 4], [4, 3]]
source = 0
destination = 5 

friends = Graph() 

for connections in friendship_circle:
    friends.add_edge(connections[0],connections[1]) 

friends = friends.save_graph() 
print("Friendship Graph : ")
display_graph(friends)
Dfs_friends(friends,source,destination)
```
