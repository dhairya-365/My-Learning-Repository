# **1. Working with a Singly linked list**
- Inserting an element at head of linked list
- Deleting an element at from linked list 
- finding the length of linked list 
- searching an element in linked list

```python
class node:
    def __init__(self,value,next=None):
        self.value = value 
        self. next = next 

    # Insertion operation 

    def insert_at_begin(self,value):
        new_node = node(value,self)
        self = new_node
        return self  

    def insert_at_end(self,value):
        new_node = node(value,None)
        tmp = self 
        while tmp.next is not None:
            tmp = tmp.next 
        tmp.next = new_node
        return self 

    def insert_at_specific_position(self,value,index):
        n = self.length() 
        if index >= n:
            self = self.insert_at_end(value) 
            return self
        elif index<0:
            self = self.insert_at_begin(value)
            return self  
        else:
            tmp = self
            count = 0
            while count != index -1 :
                # print(tmp.value)
                tmp = tmp.next
                count = count + 1 
            tmp2 = tmp.next
            tmp.next = node(value,tmp2)
        return self 

    # Deletion operation 

    def delete_at_begin(self):
        tmp = self 
        self = tmp.next 
        return self

    def delete_at_end(self):
        tmp = self 
        while tmp.next.next is not None:
            tmp = tmp.next 
        tmp.next = None 
        return self

    def delete_at_specific_position(self,index):
        n = self.length() 
        if index >= n:
            self = self.delete_at_end() 
            return self
        elif index<0:
            self = self.delete_at_begin()
            return self  
        else:
            tmp = self
            count = 0
            while count != index -1 :
                # print(tmp.value)
                tmp = tmp.next
                count = count + 1 
            tmp2 = tmp.next.next
            tmp.next = tmp2
        return self 

    def search_an_element(self,value):
        tmp = self 
        index = 0 
        while tmp is not None:
            index = index + 1 
            if tmp.value == value:
                print(f"element found at index : {index}")
                return tmp.value , index 
            tmp = tmp.next 
        print("Element not found.") 
        return None 

    # Helper function 

    def length(self):
        linkedlist = []
        while self is not None:
            linkedlist.append(self.value)
            self = self.next
        return len(linkedlist)

    def traverse(self):
        linkedlist = []
        while self is not None:
            linkedlist.append(self.value)
            self = self.next
        print(linkedlist)
```

```python
head = node(10) 
num1 = node(20)
head.next = num1
head = head.insert_at_begin(5)
head = head.insert_at_end(40)
head = head.insert_at_specific_position(50,4)
head = head.delete_at_begin()
head = head.delete_at_end()
head = head.delete_at_specific_position(2)
head.search_an_element(20)
head.traverse()
head.length()
```

# **2. Doubly LinkedList**

```python

class node:
    def __init__(self,prev,value,next):
        self.prev = prev 
        self.value = value 
        self.next = next 


    # Insertion functions

    def Insert_at_beginning(self,value):
        new_node =  node(None,value,self) 
        self = new_node
        return self
    
    def Insert_at_end(self,value):
        tmp = self 
        while tmp.next:
            tmp = tmp.next 
        
        tmp.next = node(tmp,value,None)
        return self
    
    def Insert_at_specific_position(self,value,index):
        tmp = self 
        count = 0 
        n = self.length()

        if index <= 0:
            self.Insert_at_beginning(value) 
        elif index >= n:
            self.Insert_at_end(value) 
        else:
            while count != index-1:
                tmp = tmp.next 
                count = count + 1
            tmp2 = tmp.next
            tmp.next = node(tmp,value,tmp2)
        return self  

    # Deletion operations 
    def delete_at_beginning(self): 
        self = self.next  
        self.prev = None
        return self 

    def delete_at_end(self):
        tmp = self 
        while tmp.next.next:
            tmp = tmp.next
        tmp.next = None 
        return self 

    def delete_at_specific_position(self,index):
        tmp = self 
        count = 0 
        n = self.length()

        if index <= 0:
            self = self.delete_at_beginning() 
            return self
        elif index >= n:
            self.delete_at_end() 
        else:
            while count != index-1:
                tmp = tmp.next 
                count = count + 1
            tmp2 = tmp.next
            tmp.next = tmp2.next
        return self  


    # Searching operation 

    def search_by_value(self,value):
        tmp = self 
        index = 0 
        while tmp:
            if tmp.value == value:
                print("Value found at index : ",index)
                return index
            index = index + 1 
            tmp = tmp.next 
            
        print("element not present")
        return None
        
    # helper functions 
    def traversal(self):
        linked_list = []
        tmp = self 
        while tmp:
            linked_list.append(tmp.value)
            tmp = tmp.next 
        print(linked_list)

    def length(self):
        linked_list = []
        tmp = self 
        while tmp:
            linked_list.append(tmp.value)
            tmp = tmp.next 
        return len(linked_list)
```

```python
head = node(None,5,None) 
num1 = node(head,10,None) 
num2 = node(num1,15,None)
head.next = num1
num1.next = num2 
print(f"Original Doubly Linkedlist : ")
head.traversal()

head = head.Insert_at_beginning(0)
head = head.Insert_at_end(20)
head = head.Insert_at_end(30)
head = head.Insert_at_specific_position(25,5)
print("After Insertion operations : ")
head.traversal()

head = head.delete_at_beginning()
head = head.delete_at_end()
head = head.delete_at_specific_position(0)
print("after performing deletion operations : ")
head.traversal()

head.search_by_value(20)
```

# **3. Circular linkedlist**

```python
class node:
    def __init__(self,value,next=None):
        self.value = value 
        self.next = next 

    # Insertion functions
    def Insert_at_beginning(self,value):
        tmp = self 
        while tmp.next != self:
            tmp = tmp.next 
        new_node = node(value,self) 
        tmp.next = new_node
        self = new_node
        return self

    def Insert_at_end(self,value):
        tmp = self 

        while tmp.next != self:
            tmp = tmp.next 

        tmp.next = node(value,self) 
        return self

    def insert_at_specified_position(self,value,index):
        tmp = self 
        n = self.length() 
        count = 0 
        if index <= 0:
            self = self.Insert_at_beginning(value)
        elif index >= n:
            self.Insert_at_end(value) 
        else: 
            while count!= index:
                count = count + 1 
                tmp = tmp.next 
            
            tmp2 = tmp.next 
            tmp.next = node(value,tmp2) 
        return self 

    # Deletion function 
    def delete_at_beginning(self):
        tmp = self 
        while tmp.next != self:
            tmp = tmp.next 
        tmp.next = self.next
        self = self.next  
        return self 

    def delete_at_end(self):
        tmp = self 
        while tmp.next.next != self:
            tmp = tmp.next 
        tmp.next = self 
        return self

    def delete_at_specific_position(self,index):
        tmp = self 
        n = self.length()
        count = 0
        if index <= 0:
            self = self.delete_at_beginning() 
        elif index >= n:
            self = self.delete_at_end() 
        else: 
            while count != index -1 :
                tmp = tmp.next
                count = count + 1 
            tmp2 = tmp.next.next
            tmp.next = tmp2
        return self 


    # helper function 
    def traversal_list(self):
        if self:
            tmp = self 
        linked_list = [tmp.value]
        while tmp.next != self:
            tmp = tmp.next
            linked_list.append(tmp.value) 
        return linked_list
    
    def length(self):
        if self:
            tmp = self 
        linked_list = [tmp.value]
        while tmp.next != self:
            tmp = tmp.next
            linked_list.append(tmp.value) 
        return len(linked_list)
```
