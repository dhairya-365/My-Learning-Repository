# List, tuple, set, dict 

**1. Lists : mutable**
- index access = O(1)
- Append = O(1)
- Insert/Delete = O(n)
- Search element = O(n)

```python
my_list = [10,20,30,40,"Hello world",60]
print("List before operation : ",my_list)
print("-------- Operation on list --------") 
print(my_list.append(70)) 
print(my_list)
print(my_list.pop(4))
print(my_list) 
print("accessing particular element of the list : ",my_list[3]) 
print("Slicing and indexing : ",my_list[0:2]) 
```

**2. tuple : immutable**
- index access = O(1) 
- Insertion/ Deletion = creates full copy of tuple and inserts the element, O(n) 
- search element = O(n)

```python
my_tuple = (1,2,3,4,5,6,10) 
print(my_tuple)
print("Tuple before operation",my_tuple) 
print("-------- Tuple Operations --------") 
print("inserting element into tuple by converting it into list.") 
tmp = list(my_tuple) 
tmp.append(20) 
my_tuple = tuple(tmp) 
print(my_tuple)
print("deleting element from the tuple by converting it into list.") 
tmp.pop() 
my_tuple = tuple(tmp) 
print(my_tuple)
print("accessing a particular element of the tuple.") 
print(my_tuple[2])
print("slicing operation on a tuple.") 
print(my_tuple[0:2])
```

**3. Dictionary (`dict`) - Mutable**
- index access = O(1) 
- Insert/Delete = O(1) 
- search element = O(1) 
- copying element = O(n)

```python
my_dict = {
    1 : "abc",
    2 : "def",
    3 : "ghi"
}

print(my_dict) 
print("-------- Performing operation on Dict --------") 
print("Inserting an element in dict : ") 
my_dict[4] = "jkl" 
print(my_dict) 
print("Deleting an element from the dict : ")
my_dict.pop(2) 
print(my_dict) 
print("accessing value of an element from its particular key.") 
print(my_dict[3])
```
