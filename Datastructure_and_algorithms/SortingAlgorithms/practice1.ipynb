# **1. Selection Sort**
- select the minimum element in each iteration
- Time complexity : O(N^2)
- performs swapping

```python
arr = [5,6,2,10,11,12]

def selection_sort(arr):
    n = len(arr) 
    for i in range(n-1):
        min_index = i 

        for j in range(i+1,n):
            if arr[j] <= arr[min_index]:
                min_index = j 
        
        arr[i],arr[min_index] = arr[min_index],arr[i] 
    return arr 

print(f"Array before sorting : {arr}")
print(f"Array after sorting : {selection_sort(arr)}")
```

# **2. Bubble Sort**
- sort the array from upper bound larger elements are pushed backward.
- Time complexity : O(N^2)
- performs swapping

```python
arr = [5,12,2,10,11,1] 

def Bubble_Sort(arr):
    n = len(arr) 

    for i in range(n-1,-1,-1):
        for j in range(i):
            if arr[j] > arr[j+1]:
                arr[j],arr[j+1] = arr[j+1],arr[j] 
        print(arr)
    return arr 
print(f"Array before sorting : {arr}") 
print(f"Array after sorting : {Bubble_Sort(arr)}")
```

# **3.Insertion sort**
- Select an element in each iteration and place it to its perfect place.
- Time complexity : O(N^2) 
- performs replacing

```python
arr = [10,1,7,2,6,3,0] 

def Insertion_sort(arr):
    n = len(arr)
    for i in range(1,n):
        key = arr[i] 
        j = i-1 

        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j] 
            j = j-1 

        arr[j+1] = key 
        print(arr)
    return arr 
print(f"Array before sorting : {arr}") 
print(f"Array after sorting : {Insertion_sort(arr)}")
```

# **4. Merge sort**
- works on divide and conquer princple 
- Time complexity : O(nlogn) 

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])

    return merge(left_half, right_half)

def merge(left, right):
    sorted_list = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            sorted_list.append(left[i])
            i += 1
        else:
            sorted_list.append(right[j])
            j += 1
            
    sorted_list.extend(left[i:])
    sorted_list.extend(right[j:])
    
    return sorted_list


arr = [38, 27, 43, 3, 9, 82, 10]

print(f"Array before sorting : {arr}") 
print(f"Array after sorting : {merge_sort(arr)}")
 
```

# **5. Quick Sort** 
- Works on principle of Partitoning and recursion
- Time complexity : O(n logn)

```python
arr = [38, 27, 43, 3, 9, 82, 10]

def quick_sort(arr):
    n = len(arr)

    if n <= 1:
        return arr

    pivot_element = arr[n//2] 

    left_part = [element for element in arr if element < pivot_element]
    mid = [element for element in arr if element == pivot_element]
    right_part = [element for element in arr if element > pivot_element]
    return quick_sort(left_part) + mid + quick_sort(right_part) 

print(f"Array before sorting : {arr}") 
print(f"Array after sorting : {quick_sort(arr)}")
```
