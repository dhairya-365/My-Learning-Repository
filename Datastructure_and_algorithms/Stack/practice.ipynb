# 1. Stack - LIFO : Last In First Out

```python
stack = [10,20,30,40] 

def insert(element,stack):
    stack.append(element)
    top = len(stack)
    return stack,top 

def remove(stack):
    element = stack.pop()
    return stack, element
    
```

```python
print(insert(50,stack))
print(remove(50,stack))
```
