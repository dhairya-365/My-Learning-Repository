# Recursive Backtracking
#### **1. The "Pure Generation" Problem** :

**The Goal**: Given an integer $N$, generate all possible binary strings (combinations of 0s and 1s) of length $N$.\
**Example**: If $N = 2$, output ["00", "01", "10", "11"].

```python
def generate_binary(num):
    sol = []
    def backtrack(curr_path):

        # base case 
        if len(curr_path) == num:
            sol.append("".join(curr_path)) 
            return 
        
        # choice-1 : choose 0
        curr_path.append("0")
        backtrack(curr_path) 
        curr_path.pop() 

        # choice-2 : choose 1 
        curr_path.append("1")
        backtrack(curr_path) 
        curr_path.pop() 
    
    backtrack([])
    return sol 
```

```python
print(generate_binary(4))
```

#### **2.The "Pruning" Problem: Combination Sum**

**The Goal**: Given a list of positive numbers and a target sum, find all unique combinations of numbers that add up exactly to that target. You can use a number more than once.\
**Example**: nums = [2, 3, 6, 7], target = 7. Output: [[2, 2, 3], [7]].

```python
def Combination_sums(nums,target):
    result = []
    def backtrack(start_idx,sol):
        # Base case 
        if sum(sol) == target:
            result.append(sol[:]) 
            return 
        if sum(sol) > target:
            return 

        # Choice-1 pick the number
        for i in range(start_idx,len(nums)):
            start_idx =  i
            sol.append(nums[i])
            backtrack(start_idx,sol)
            sol.pop() 

    backtrack(0,[])
    return result
```

```python
Combination_sums([1,2,3],target=6)
```

#### **3. The "Mapping" Problem: Phone Keypad**

**The Goal**: Think of an old mobile phone where 2 is "abc", 3 is "def", etc. Given a string of digits (like "23"), return all possible letter combinations that the numbers could represent.\
**Example**: digits = "23". Output: ["ad","ae","af","bd","be","bf","cd","ce","cf"].

```python
def keypad(digits):
    mapping = {
        "2" : "abc",
        "3" : "def"
    }
    results = []

    def backtrack(start_idx,sol):

        # Base cases 
        if len(sol) == len(digits):
            print(sol)
            results.append("".join(sol[:])) 
            return 
        if len(sol)>len(digits):
            return 
        
        # choices
        curr_digit = digits[start_idx] 
        available_num = mapping[curr_digit] 

        for num in available_num:
            sol.append(num) 
            backtrack(start_idx+1,sol) 
            sol.pop()
        
    backtrack(0,[])
    return results 
```
