# 1. Inheritance

**MultiLevel inheritance**

```python
class parent_class:
    def __init__(self,name,age):
        self.name = name
        self.age = age 

    def print_info(self):
        print(self.name,self.age)

class parent_class2(parent_class):
    def __init__(self,name,age):
        super().__init__(name,age)
        pass 
    
    def print_info(self):
        print("class 2 ")
        print(self.name,self.age)

class child_class(parent_class2):
    pass


x = child_class("Dhairya",22) 
print(x.print_info())
```

**Hierarchical  Inheritance**

```python
class parent_class1:
    def __init__(self,name,age):
        self.name = name 
        self.age = age 
    
    def print_employee_info(self):
        print("Name : ",self.name,"Age : ",self.age)
    

class child_class1(parent_class1):
    def __init__(self,name,age):
        super().__init__(name,age) 
        self.name = name 
    
    def welcome_employee(self):
        print("Welcome, ",self.name) 

class child_class2(parent_class1):
    def __init__(self,name="Unknown",age=18):
        super().__init__(name,age)
    def user_age(self):
        print("age : ",self.age)

obj = child_class2(18)
obj.user_age()
```

**Multiple Inheritance**

```python
class Arithmetic_operation_class:
    def __init__(self,num1,num2):
        self.num1 = num1 
        self.num2 = num2 
    
    def addition(self):
        return self.num1 + self.num2 
    
    def subtraction(self):
        return self.num1 - self.num2 
    
    def multiplication(self):
        return self.num1 * self.num2 
    
    def division(self):
        return self.num1 / self.num2 


class logical_operation_class:
    def __init__(self,value1,value2):
        self.value1 = value1 
        self.value2 = value2 
    
    def AndOperation(self):
        return self.value1 & self.value2 
    
    def OrOperation(self):
        return self.value1 ^ self.value2 
    

class derived_class(Arithmetic_operation_class,logical_operation_class):
    def __init__(self,num1=None,num2=None,value1=None,value2=None):
        Arithmetic_operation_class.__init__(self,num1,num2) 
        logical_operation_class.__init__(self,value1,value2) 

obj = derived_class(num1=10,num2=20) 
print(obj.addition())
obj2 = derived_class(value1=True,value2=False) 
print(obj2.AndOperation())
```

# 2. Polymorphism

```python
class Vehicle:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

class Car(Vehicle):
  def move(self):
    print("Moving")

class Boat(Vehicle):
  def move(self):
    print("Sailing")


car_obj = Car("Porsche","GT3s") 
boat_obj = Boat("Bayliner","Bowrider")  

car_obj.move() 
boat_obj.move()
```

# 3. Data Encasulation 

```python
class parent_class:
    def __init__(self,area):
        self.area = area 
        self._shape = "Cicle"
        self.__pie = 3.14 

    def calculate_area(self):
        print("area of circle :- ",self.area * self.__pie) 

class child_class(parent_class):
    def __init__(self,area):
        super().__init__(area)
        print(self.area) # public
        print(self._shape) # protected
        print(self.__pie) # private


obj = child_class(10) 
obj.calculate_area() 
```

# 4. Data Abstraction 

```python
from abc import ABC,abstractmethod 

class Details(ABC):
    @abstractmethod
    def employee_details(self,name:str,Uid:str):
        pass 

class employee(Details):
    def __init__(self): 
     pass 

    def employee_details(self,name:str,Uid:str):
        print("Employee details") 
        print("Name : ",name)
        print("Uid : ",Uid)
    
class owner(Details):
    def __init__(self):
        pass
    
    def employee_details(self,name,Uid,salary):
        print("Employee details") 
        print("Name : ",name)
        print("Uid : ",Uid)
        print("salary : ",salary)


obj_employee_class = employee()
obj_owner_class = owner() 

obj_employee_class.employee_details(name="Dhairya",Uid="D001")
obj_owner_class.employee_details(name="Dhairya",Uid="D001",salary=10000)
```
